```python
# Mapping of organs to their cancer abbreviations
remap_dict = {
    "Pancreas": "PDAC",
    "Skin": "SKCM",
    "Uterus": "UCEC",
    "Ovary": "OV",
    "Breast": "BRCA",
    "Bone_marrow": "MM",
    "Brain": "GBM",
    "Kidney": "ccRCC",
    "Head_neck": "HNSCC",
    "Cervical": "CESC",
    "Colorectal": "CRC"
}

def reclassify(source):
    if "Normal" in source:
        return source  # Keep normals as-is
    organ, tumor_type = source.split("_", 1)
    if organ in remap_dict:
        prefix = remap_dict[organ]
        if tumor_type == "Primary tumor":
            return f"{prefix}_PT"
        elif tumor_type == "Met":
            return f"{prefix}_Met"
    return source  # Leave unchanged if no match

# Apply transformation
adata.obs["Source_type_v2"] = adata.obs["Source_type"].apply(reclassify)
```

```python
adata.obs["Source_type_v2"] = adata.obs["Source_type_v2"].replace({
    "Bone_marrow_Primary tumor": "MM_PT",
    "Bone_marrow_Met": "MM_Met",
    "Head_neck_Primary tumor": "HNSCC_PT",
    "Head_neck_Met": "HNSCC_Met",
    "Blood_Normal": "PBMC"
})
```

```python
# Replace all "_Normal" suffixes with "_N"
adata.obs["Source_type_v2"] = adata.obs["Source_type_v2"].str.replace("_Normal$", "_N", regex=True)
```
