# Addition of Data.source_Chemistry column to the metadata

```python
import pandas as pd

# Load the CSV file into a DataFrame
df = pd.read_csv('meta_CLEAN_nkcells_RNA.csv')

# Concatenate the two columns into a new column
df['Data.source_Chemistry'] = df['Data.source'] + "_" + df['Chemistry']

# Save the updated DataFrame to a new CSV
df.to_csv('meta_CLEAN_nkcells_RNA_DataSource_Chemistry.csv', index=False)

print("Concatenation complete. Output saved to 'meta_updated.csv'")
df['Data.source_Chemistry']
```
