```python
# Load dataset paths and RNA counts
dataset_paths = [
    '/Volumes/queries/PCIML2/PCIML2_eMLclassified_adata.h5ad',
    '/Volumes/queries/PCIML7/PCIML7_eMLclassified_adata.h5ad',
    '/Volumes/queries/PCIML8/PCIML8_eMLclassified_adata.h5ad',
    '/Volumes/queries/TCIML2/TCIML2_eMLclassified_adata.h5ad',
    '/Volumes/queries/TCIML3/TCIML3_eMLclassified_adata.h5ad',
    '/Volumes/queries/TCIML7/TCIML7_eMLclassified_adata.h5ad',
]

rna_counts_files = {
    'PCIML2': '/Volumes/queries/PCIML2/PCIML2_RNAcounts.csv',
    'PCIML7': '/Volumes/queries/PCIML7/PCIML7_RNAcounts.csv',
    'PCIML8': '/Volumes/queries/PCIML8/PCIML8_RNAcounts.csv',
    'TCIML2': '/Volumes/queries/TCIML2/TCIML2_RNAcounts.csv',
    'TCIML3': '/Volumes/queries/TCIML3/TCIML3_RNAcounts.csv',
    'TCIML7': '/Volumes/queries/TCIML7/TCIML7_RNAcounts.csv',
}

probability_columns = ['CD56brightBBCprob', 'CD56dimBBCprob', 'eML1BBCprob', 'eML2BBCprob']
labels = ['CD56bright', 'CD56dim', 'ML1', 'ML2']

def classify_row(row):
    if all(row[col] < 0.5 for col in probability_columns):
        if row['eML1BBCprob'] + row['eML2BBCprob'] > 0.5:
            return 'ML_transition'
        else:
            return 'unclassified'
    else:
        max_index = row[probability_columns].astype(float).idxmax()
        return labels[probability_columns.index(max_index)]

# Process each dataset
classified_datasets = []

for path in dataset_paths:
    dataset_name = Path(path).parent.name
    
    print(f"\n{'='*60}")
    print(f"Processing: {dataset_name}")
    print(f"{'='*60}")
    
    # Load dataset
    adata = sc.read_h5ad(path)
    print(f"Loaded: {adata.n_obs:,} cells × {adata.n_vars:,} genes (subset)")
    
    # Filter for Donor(Day zero)
    donor_mask = adata.obs['orig.ident'].str.contains('Donor', na=False)
    adata = adata[donor_mask].copy()
    print(f"After donor filter: {adata.n_obs:,} cells")
    
    if adata.n_obs == 0:
        print(f"⚠️  Skipping - no donor cells")
        continue
    
    # ADD FULL RNA COUNTS HERE 
    rna_path = rna_counts_files[dataset_name]
    print(f"\nAdding full RNA counts from CSV...")
    
    # Load RNA counts
    counts_df = pd.read_csv(rna_path, index_col=0).T  # Transpose to cells × genes
    print(f"  RNA counts: {counts_df.shape[0]} cells × {counts_df.shape[1]} genes")
    
    # Convert barcode format: RNA has .1, adata has -1
    def rna_to_adata_barcode(bc):
        return bc.replace('.1', '-1')
    
    counts_df.index = [rna_to_adata_barcode(bc) for bc in counts_df.index]
    
    # Find matching cells
    common_cells = adata.obs_names.intersection(counts_df.index)
    print(f"  Matched: {len(common_cells)} cells")
    
    if len(common_cells) == 0:
        print(f"  ⚠️  No matches - skipping")
        continue
    
    # Filter both to common cells
    adata = adata[common_cells, :].copy()
    counts_subset = counts_df.loc[common_cells, :]
    
    # Create new AnnData with full genes + existing metadata
    adata_full = sc.AnnData(
        X=counts_subset.values,
        obs=adata.obs,
        var=pd.DataFrame(index=counts_subset.columns)
    )
    
    print(f"  ✓ Updated: {adata_full.n_obs} cells × {adata_full.n_vars} genes (FULL)")
    
    # Apply classification
    for col in probability_columns:
        adata_full.obs[col] = pd.to_numeric(adata_full.obs[col], errors='coerce')
    
    adata_full.obs['classification'] = adata_full.obs.apply(classify_row, axis=1)
    adata_full.obs['dataset'] = dataset_name

    #ADDING DATASET PREFIX ONLY FOR TCIML7_Donor as it lacks identifier and again it will be seen as duplicated barcode with other barcodes
    if dataset_name == 'TCIML7':
        def add_tciml7_prefix(row):
            """Add TCIML7_ prefix only for TCIML7_Donor cells"""
            if row['orig.ident'] == 'TCIML7_Donor':
                barcode = row.name
                if not barcode.startswith('TCIML7_'):
                    return f"TCIML7_{barcode}"
            return row.name
        
        adata_full.obs_names = adata_full.obs.apply(add_tciml7_prefix, axis=1)
        print(f"  ✓ Added TCIML7_ prefix to donor cells")
    
    print(f"  Sample barcodes: {adata_full.obs_names[:3].tolist()}")

    print(f"  Classification:")
    print(adata_full.obs['classification'].value_counts())
    
    classified_datasets.append(adata_full)

print(f"\n{'='*60}")
print(f"Processed {len(classified_datasets)} datasets")
print(f"{'='*60}")

# NOW combine with full genes
combined_adata = sc.concat(
    classified_datasets,
    join='outer'
)

print(f"\n✅ FINAL COMBINED:")
print(f"Cells: {combined_adata.n_obs:,}")
print(f"Genes: {combined_adata.n_vars:,}")
print(f"Total counts in first cell: {combined_adata.X[0].sum():.0f}")

print(f"\nClassification:")
print(combined_adata.obs['classification'].value_counts())
```
